```python
from sklearn import model_selection
```

```python
import numpy as np
```

```python
X = 2*np.random.rand(100,1)

y = 3*X +np.random.rand(100,1)+2

```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.ylim((0,14));
```

```python
X_b = np.c_[np.ones((100,1)),X]
X_b[:10]
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_b.T.dot(X_b)
```

```python
np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
plt.scatter(X,y)
plt.ylim((0,14));
plt.plot(X_b[2],theta_best)
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]# add x0 = 1 to each instance
print(X_new_b)
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict,"r-")
plt.scatter(X, y)
plt.axis((0,2,0 ,15))
plt.show()
```

```python
plt.plot((1,3))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lnr_reg = LinearRegression()
lnr_reg.fit(X,y)
```

```python
lnr_reg.coef_,lnr_reg.intercept_
```

```python
x_test = np.array([[2],[7]])
test = np.c_[np.ones((2,1)),x_test]
```

```python
test
```

```python
test_data = test.dot(theta_best)
test_data
```

```python
plt.plot(x_test,test_data)
```

```python
preds=lnr_reg.predict(x_test)
```

```python
lnr_reg.score(preds,test_data)
```
