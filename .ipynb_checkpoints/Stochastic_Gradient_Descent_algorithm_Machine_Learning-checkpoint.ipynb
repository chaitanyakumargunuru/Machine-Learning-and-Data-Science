# Stochastic Gradient Descent Algorithm of Machine Learning
> *Stochastic* means Randomness , so in this algorithm ,model selects random feature then calculate Gradient for *global*Minimum .For this model to work better `LearningRate(eta0)` is important feature for hyperparameter tuning .It works better than `LinearRegression` model due to its *Out-of-core* method it trains one feature at a time while former train whole dataset at a time so it will take more duration for training the large dataset.


* Now, coming to the dataset we are working is from kaggle :https://www.kaggle.com/datasets/yasserh/walmart-dataset
* Dataset has following columns/features : 
  * Date - the week of sales
  * Weekly_Sales - sales for the given store
  * Holiday_Flag - whether the week is a special holiday week 1 – Holiday week 0 – Non-holiday week
  * Temperature - Temperature on the day of sale
  * Fuel_Price - Cost of fuel in the region
  * CPI – Prevailing consumer price index
  * Unemployment - Prevailing unemployment rate

Holiday Events<br /> Super Bowl: 12-Feb-10, 11-Feb-11, 10-Feb-12, 8-Feb-13<br /> Labour Day: 10-Sep-10, 9-Sep-11, 7-Sep-12, 6-Sep-13<br /> Thanksgiving: 26-Nov-10, 25-Nov-11, 23-Nov-12, 29-Nov-13<br /> Christmas: 31-Dec-10, 30-Dec-11, 28-Dec-12, 27-Dec-13

## Preparing the data or Data cleaning 
Preparation of data involves:
* Data imputing (filling the missed data or removing the data instance where data is missing)"ofcourse here we don't require this".
* Data Encoding (One_hot encoding the data)
* Normalization or Standardization
* ColumnTrannsformation

```python
# let's check if data contains null/NA data
data.isna().sum()
```

### Here we have the date column in object so we one_hot_encode it
### Before we going to it first we have to standardize the data

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score , root_mean_squared_error,mean_absolute_error
from sklearn.linear_model import SGDRegressor
```

```python
from sklearn.model_selection import train_test_split
X,y = data.drop("Weekly_Sales",axis=1),data["Weekly_Sales"]
X_train,X_test,y_train, y_test = train_test_split(X,y,
                                                   random_state = 123,
                                                   test_size = 0.1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler , OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
X_train
```

```python
pipe = ColumnTransformer([("scale",StandardScaler(),["Store","Temperature","Fuel_Price","CPI","Unemployment"]),("encode",OrdinalEncoder(),["Date"])],remainder = "passthrough")
```

```python
transformed_train_data = pipe.fit_transform(X_train)
```

```python
transformed_train_data
```

```python
transformed_test_data = pipe.fit_transform(X_test)
```

```python
transformed_test_data[0] # we have data column in still numerical but not in normalized 
```

```python
scale = StandardScaler()
transformed_train_data = scale.fit_transform(transformed_train_data)
```

```python
transformed_train_data
```

```python
transformed_test_data = scale.fit_transform(transformed_test_data)
```

## Let's create and Train the model

```python
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor()
g=sgd.fit(transformed_train_data,y_train)
```

```python
sgd.score(transformed_train_data,y_train)
```

## Evaluating and FineTuning

```python
y_preds = sgd.predict(transformed_test_data)
```

```python
r2_score(y_test,y_preds)
```

```python
root_mean_squared_error(y_test,y_preds)
```

```python
sgd = SGDRegressor(max_iter = 100000)
g=sgd.fit(transformed_train_data,y_train)
```

```python
sgd.score(transformed_train_data,y_train)
```

```python
plt.scatter(np.arange(len(y_test)),y_test,label="test")
plt.scatter(np.arange(len(y_preds)),y_preds,label="preds")
plt.legend();
```

```python
len(y_preds),len(y_test)
```

```python
mean_absolute_error(y_test,y_preds)
```

```python
from sklearn.linear_model import LinearRegression
lnr = LinearRegression()
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =3 ,include_bias = False)
x_poly = poly.fit_transform(x_train)
```

```python
lnr.fit(x_poly,y_train)
```

```python
lnr.score(x_poly,y_train)
```

```python
x_poly_test = poly.fit_transform(x_test)
lnr.score(x_poly_test,y_test)
```
