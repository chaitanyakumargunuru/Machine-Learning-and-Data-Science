## DATA VISUALIZATION OF CAR_SALES.CSV FILE USING MATPLOTLIB

```python
fig,ax=plt.subplots(nrows=2,ncols=3,figsize=(40,20))
ax[0,0].bar(car_sales.Make,car_sales.Price);
ax[0,0].set(title="car sales",xlabel="Make",ylabel="Price");
ax[0,1].scatter(car_sales.Make,car_sales["Odometer (KM)"])
ax[0,1].set(xlabel="Make",ylabel="Odometer (KM)");
ax[0,2].plot(car_sales["Sale Date"],car_sales.Price.cumsum())
ax[0,2].set(xlabel="Sale Date",ylabel="Car price");
ax[1,0].plot(car_sales["Odometer (KM)"],car_sales["fuel used (L)"])
ax[1,0].set(xlabel="Odometer ",ylabel="fuel used");
ax[1,1].scatter(car_sales.Make,car_sales["Sale Date"])
ax[1,1].set(xlabel="MAKE",ylabel="sale date");
ax[1,2].barh(car_sales.Make,car_sales.fuel_eco)
ax[1,2].set(xlabel="make",ylabel='fuel eco');
fig.savefig("images/car_sales.png")
```

## we are using two methods of plotting , which one is better
* for faster plotting use pyplot
* for advanced and detailed use matplotlib OO method

```python
hd.head()
```

```python
over_50=hd[hd.age>50]
over_50.tail()
```

```python
#pyplot
over_50.plot(kind="scatter",figsize=(10,10),x="age",y="chol",c="target");
```

```python
# using matplotlib oo method mixed with pyplot
fig,ax=plt.subplots(figsize=(10,8))
over_50.plot(kind="scatter",x="age",y="chol",c='target',ax=ax,legend='reverse'
             ,grid=bool,fontsize=20,colorbar="red");
ax.set_xlim([50,80]);
```

## pure oo method

```python

fig,ax=plt.subplots(figsize=(10,8))
#plotting the data
scatter=ax.scatter(x=over_50.age,y=over_50.chol,c=over_50.target) 
#axes setting title and labeling
ax.set(xlabel="age",ylabel="cholestrol",title="Heart disease and cholesterol levels",facecolor="cyan");
#setting legend
ax.legend(*scatter.legend_elements(),title="target",facecolor="grey",fontsize=10,labelcolor="cyan");
#plotting mean line
ax.axhline(hd.chol.mean(),linestyle="--",color="orange");
```
